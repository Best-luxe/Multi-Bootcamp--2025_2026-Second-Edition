## create a program that calculates the volume and surface area of a cube based on the lenghth of one side provided by the user.

```python

length=float(input("Enter your length"))
height=float(input("Enter your height"))
weight=float(input("Enter your weight"))
volume=length*height*weight
surface_area=2*(length*weight+weight*height+length*height)
print(f"the volume of a cube is {volume} and the surface area of the cube={surface_area}")
```

## Write a program that takes a base salary and calculates net pay after deducting taxes (assume a fixed tax rate of 20%).

```python
base_salary=float(input("Enter base_salary:"))
tax_rate=20
tax_amount=(tax_rate/100)*base_salary
net_pay=base_salary-tax_amount
print("tax deducted:",tax_amount)
print("net pay after tax:",net_pay)
```
