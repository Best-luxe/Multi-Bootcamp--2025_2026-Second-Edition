## Create a function 
calculate_discount that takes a price and discount percentage (default 10%) and return the discounted price.

```python
def calculate_discount(price,discount_percentage=10):
    discounted_price=price-(price*discount_percentage/100)
    return discounted_price
```

```python
print(calculate_discount(100)) #uses default10% 90.0
print(calculate_discount(200,20)) #20% discount 160.0
```

## QUESTION 8

## Define a function create_email that takes a first name,last name,and optional domain (default "gmail.com") to create  an email address.

```python
def create_email(first_name,last_name,domain="gmail.com"):
    email=f"{first_name.lower()}.{last_name.lower()}@{domain}"
    return email
```

```python
print(create_email("Best", "Naza"))
# best.naza@gmail.com
print(create_email("Kosi", "Best", "yahoo.com"))
# Kosi.best@yahoo.com
```

## QUESTION 9

## Create a function count_vowels that takes a string and returns the number of vowels in it.

```python
def count_vowels(string):
    vowels="aeiou"
    count=0
    for char in string:
        if char in vowels:
            count+=1
            return count
```

```python
print(count_vowels("hello world"))
```

## QUESTION 12

## Write a funtion file_extension that extracts the file extension from a filename,handling cases where there there might be no extension.

```python
def file_extension(filename):
   if "." in filename:
       return filename.split(".")[-1]
   else:
       return "No extension"
```

```python
print(file_extension("data.csv")) # csv
print(file_extension("file")) # No extension
```

## Write a funtion safe_get_item that tries to access a list item by index and returns none if the index is out of bounds.

```python
def safe_get_item(list,index):
    if 0<=index<len(list):
        return list[index]
    else:
        return None
```

```python
numbers=[10,20,300]
print(safe_get_item(numbers,1)) # 20
print(safe_get_item(numbers,5)) #None
print(safe_get_item(numbers,-1)) # None
```

## Create a funtion clean_data that takes a list of mixed data types and returns only the numeric values as a new list.

```python
def clean_data(data_list):
    numeric_values=[]
    for item in data_list:
        if isinstance(item, (int, float)):
            numeric_values.append(item)
            return numeric_values
```

```python
mixed_data=[10,"hello", 3.5, None,True,"42", 7]
print(clean_data(mixed_data))
# output: [10,3.5,7]
```
