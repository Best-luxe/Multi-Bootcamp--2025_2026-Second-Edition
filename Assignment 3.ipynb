## QUESTION 3

## Given a list of numbers,find and  print the largest and smallest values.

```python
numbers=[4,7,1,9,2]
smallest=numbers[0]
largest=numbers[0]
for num in numbers:
    if num < smallest:
        smallest=num
        if num < largest:
            largest=num
print("smallest:",smallest)
print("largest:",largest)
```

## Write a program that removes all duplicate values from a list and prints the unique elements.

```python
numbers=[2,3,4,5,9]
unique_numbers=[]
for num in numbers:
    if num not in unique_numbers:
        unique_numbers.append(num)
print("unique element:", unique_numbers)
```

## QUESTION 5

## Write a program that finds the most common element in a list and how many times it appears.

```python
numbers=[1,3,2,3,4,3,2,3]
counts={}
for num in numbers:
    if num in counts:
        counts[num]+=1
    else:
        counts[num]=1
    most_common=max(counts, key=counts.get)
print("most common element:", most_common)
print("it appears", counts[most_common], "times")
```

## QUESTION 6

## Create a program that counts the number of vowels and consonants in a sentance

```python
sentence=input("Enter a sentence:").lower()
vowels="aeiou"
vowel_count=0
consonant_count=0
for char in sentence:
   if char.isalpha(): # only count letters
       if char in vowels:
         vowel_count+=1
else:
            consonant_count+=1
print("vowels:", vowel_count)
print("consonants:",consonant_count)
        
```

## QUESTION 7

## Write a program that finds the longest word in a sentence entered by user

```python
sentence=input("Enter a sencetence:") 
words=sentence.split() # split sentence into word
longest_word=""
for word in words:
    if len(word)>len(longest_word):
       longest_word=word
print("longest word:",longest_word)
print("length:", len(longest_word))
```

## QUESTION 12

## Create a program that prints the fibonacci sequence up to a specific count provided by the user.

```python
number_terms=int(input("Enter the number of terms: "))
fib_sequence =[0,1]
for i in range(2,number_terms):
    next_term = fib_sequence[i-1]+ fib_sequence[i-2]
    fib_sequence.append(next_term)
print( fib_sequence)
```

## QUESTION 15

## Write a program that keeps asking for a number until the user enter 0, then outputs the sum off all entered numbers.

```python
total=0
while True:
    num=int(input("Enter a number (0 to stop):"))
    if num==0:
        break
        total+=num
print("the sum of the entered numbers is:",total)
```
