# Analying a stock portfolio using numpy's mathematical function

```python
import numpy as np 
np.random.seed(789)
days=252
stocks=5
daily_return=np.random.normal(0.001,0.02, (days,stocks))
stock_names=["AAPL", "GOOGL", "MSFT", "TSLA", "AMZN"]
initial_prices=np.array([150,2800,300,800,3200])
# best analysis
cumulative_returns = np.cumprod(1+daily_returns, axis=0)-1
final_returns=cumulative_returns[-1]
# annualized volatility
volatility=np.std(daily_returns, axis=0)*np.sqrt(252)
#best and worse performers
best_idx=np.argmax(final_returns)
worse_idx=np.argmin(final_returns)
print("final cumulative returns:")
for i in range(stocks):
    print(f"{stock_names[i]}: {final_returns[i]:.2%}")
print(f"/nbest performing stock: {stock_names[best_idx]}")
print(f"/nbest performing stock: {stock_names[worse_idx]}")
```

# Risk metrics

```python
# sharpe rotio (assuming risk-free rate = 0)
mean_returns=np.mean(daily_returns,axis=0)*252
sharpe_ratio=mean_returns/volatility
print("/nsharpe ratios:")
for i in range(stocks):
    print(f"{stock_names[i]}: {sharpe_ratio[i]:.2f}")
```

```python
# maximum drawdown
def max_drawdown(returns):
    cumulative=np.cumprod(1+returns)
    peak=np.maximum.accumulate(cumulative)
    drawdown=(cumulative-peak)/peak
    return np.min(drawdown)
print("/nmaximum drawdown:")
for i in range(stocks):
    mdd=max_drawdown(daily_returns[:,i])
    print(f"{stock_names[i]}: {mdd:2%}")
```

```python
# correlation matrix
correlation_matrix=np.corrcoef(daily_returns.T)
print("/ncorrelation matrix:")
print(correlation_matrix)
```
